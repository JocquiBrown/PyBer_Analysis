```python
%matplotlib inline

# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics

# Load in csv
pyber_ride_df = pd.read_csv("PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Plot the DataFrame
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")

# Add ticks for each month in our DataFrame
# Create an array for each month in the DataFrame
x_axis = np.arange(len(pyber_ride_df))

# Set ticks for each month using list comprehension
tick_locations = [value for value in x_axis]

# Plot the ticks on the x-axis
plt.xticks(tick_locations, pyber_ride_df["Month"])
```

```python
pyber_ride_df.plot.bar(x = "Month", y = "Avg. Fare ($USD)")
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
```

```python
# Obtain the standard deviation of the Avg. Fare Column
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])

# Create a bar chart of the pyber_ride_df DataFrame with capped error bars.Change the color to sky blue 
pyber_ride_df.plot(x = "Month", y = "Avg. Fare ($USD)", kind = 'bar', color = "xkcd:sky blue", yerr = stdev, capsize = 3)

# Set the y-axis increment to every $5 (Create an array from 0 to 50 with steps equal to 5) 
plt.yticks(np.arange(0, 51, step = 5.0))

# Rotate the labels on the x-axis to horizontal
plt.xticks(rotation = 0)
```
