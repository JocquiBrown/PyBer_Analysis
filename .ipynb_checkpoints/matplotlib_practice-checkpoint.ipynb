```python
%matplotlib inline
```

```python
# Import Dependencies
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() condensed
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() separated part 1
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt() separated part 2
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker = "*", color = "red", linewidth = 2, label = 'Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()

# Add a grid
plt.grid()
```

```python
# Create the plot using the OOI method
fig, ax = plt.subplots()

# Add a label for the legend, make the marker diamond, make the line width 2 points, and make the line color green
ax.plot(x_axis, y_axis, label = "Boston", marker = "D", color = "green", linewidth = 2)

# Create a title, and label the axes
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")

# Create a y-axis limit
ax.set_ylim(0, 45)

# Create a grid and a legend
ax.grid()
ax.legend()
```

```python
# Create a bar chart
plt.bar(x_axis, y_axis)
```

```python
# Create a bar chart
plt.bar(x_axis, y_axis, color = "green", label = "Boston")

# Create labels for the x and y axes and a title
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")

# Add a legend
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create a horizontal bar chart, change the color to magenta and add a label for Boston
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")

# Add a title and axes labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Create a legend
plt.legend()

# Invert the y-axis so January is at the top
plt.gca().invert_yaxis()
```

```python
# Create a bar chart with ax.plt() (OOM)
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a horizontal bar chart using the OOI Method
fig, ax = plt.subplots()

# Set the legend label to Chicago, change the bar color to cyan, and switch the Fare data to the x-axis
ax.barh(x_axis, y_axis, label = "Chicago", color = "cyan")

# Add a title and labels for the x and y axes
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Invert the y axis data
ax.invert_yaxis()

# Add the legend 
ax.legend()
```

```python
# Create a scatter plot using plt.plot()
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot using plt.scatter(), Change the color to red, Switch the axes, and add a legend for chicago
plt.scatter(y_axis, x_axis, label = "Chicago", color = "red")

# Add a title and axes labels
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBerFare by Month")

# Add a legend
plt.legend()

# Add a limit for the x-axis data
plt.xlim(0, 50)

# Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()

```

```python
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
# Create a scatter plot with larger bubbles
# Create a new list for our multiplied values
y_axis_larger = []

# Iterate through every value in the y_axis list
for data in y_axis:
    # Add the multiplied data to our new list
    y_axis_larger.append(data*3)

# Create the new bubble chart
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# Use the list comprehension technique to create the enlarged bubbles
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
# Create a scatter plot using the OOI Method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
# Using the OOIM: Create a scatter plot with sky blue markers, with 5x size bubbles, 
# black edges around the circles, with 20% transparency, with a linewidth of 2 points, 
# a legend for Boston, and switch the axes
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i*5 for i in y_axis], color = "xkcd:sky blue", 
           edgecolors = "black", label = "Boston", alpha = 0.8, linewidth = 2)

# Add a title and a legend
ax.set_title("PyBer Fare by Month")
ax.legend()

# Label the axes
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Add a limit for the x-axis data
ax.set_xlim(0, 50)

# Invert the y-axis so that January is on top
ax.invert_yaxis()
```

```python
# Create a pie chart using the MATLAB method
plt.pie(y_axis, labels = x_axis)
```

```python
# Use the explode parameter to pull out the July slice
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)

# Add percentages to our chart and increase the size of the chart
plt.subplots(figsize = (8, 8))
plt.pie(y_axis, labels = x_axis, explode = explode_values, autopct = "%.1f%%")
```

```python
# Assign a new color to each month
colors = ["slateblue", "magenta", "lightblue", "green", 
          "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))

plt.pie(y_axis, labels = x_axis, explode = explode_values, colors = colors, autopct = "%.1f%%", shadow = True, startangle = 82)
```

```python
# Create a pie chart using the OOI Method
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
```

```python
# Using OOI: increase the figure size to 8x8
fig, ax = plt.subplots(figsize = (8, 8))

# Explode the 2 months with the largest percentage of Fare
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Create a list of new color values
new_colors = ["slateblue", "magenta", "lightblue", "green", 
          "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Make sure to include the following: New Colors, a shadow, explode values, percentages to one decimal place
plt.pie(y_axis, labels = x_axis, autopct = "%.1f%%", colors = new_colors, explode = explode_values, shadow = True, 
        startangle = 263)

# Reverse the order so that months are in clockwise order
ax.invert_yaxis()
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Plot errorbars to a line chart
plt.errorbar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

```python
# Create a line chart with capped error bars using the OOI method
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

```python
# Create a bar chart with capped error bars using the MATLAB method
plt.bar(x_axis, y_axis, yerr = stdev, capsize = 3)
```

```python
# Create a bar chart with capped error bars using the OOI method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, capsize = 3, yerr = stdev)
```

```python
# Create a horizontal bar chart with ticks at intervals of 5 and increase the figsize to 8x8.
fig, ax = plt.subplots(figsize = (8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step = 5.0))

# Invert the y-axis so that January is on top
ax.invert_yaxis()

# Import MultipleLocator from matplotlib.ticker
from matplotlib.ticker import MultipleLocator

# Set minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
```
